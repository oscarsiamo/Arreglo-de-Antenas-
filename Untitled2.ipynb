```python
def convertir_coordenadas(coordenada):
    # Extraer los componentes de grados, minutos y segundos utilizando expresiones regulares
    patron = r'(\d+) (\d+)' + r"' (\d+\.\d+)" + r'"'
    resultado = re.match(patron, coordenada)
    grados = int(resultado.group(1))
    minutos = int(resultado.group(2))
    segundos = float(resultado.group(3))

    # Calcular el valor decimal
    valor_decimal = grados + minutos / 60 + segundos / 3600

    return valor_decimal
```

```python
def graficar_ovalo_inclinado(centro_latitud, centro_longitud, semieje_mayor_grados, semieje_menor_grados, angulo_inclinacion_grados, color,transparencia):
    # Convertir el ngulo de inclinacin de grados a radianes
    angulo_inclinacion_rad = np.radians(angulo_inclinacion_grados)

    # Crear una malla de puntos para el valo sin inclinacin
    num_puntos = 100
    theta = np.linspace(0, 2*np.pi, num_puntos)
    ovalo_latitud_sin_inclinacion = semieje_mayor_grados * np.cos(theta)
    ovalo_longitud_sin_inclinacion = semieje_menor_grados * np.sin(theta)

    # Aplicar la transformacin de rotacin al valo
    matriz_rotacion = np.array([[np.cos(angulo_inclinacion_rad), -np.sin(angulo_inclinacion_rad)],
                                [np.sin(angulo_inclinacion_rad), np.cos(angulo_inclinacion_rad)]])
    ovalo_latitud, ovalo_longitud = np.dot(matriz_rotacion, np.array([ovalo_latitud_sin_inclinacion, ovalo_longitud_sin_inclinacion]))

    # Trasladar el valo al centro deseado
    ovalo_latitud += centro_latitud
    ovalo_longitud += centro_longitud
    plt.fill(ovalo_longitud, ovalo_latitud, alpha=transparencia,c=color)
```

```python
# Convertir el arreglo NumPy en una serie de pandas
latitud_serie = pd.Series(df_RFI[:, 0])
longitud_serie = pd.Series(df_RFI[:, 1])

```

```python
# Definir la funcin de conversin
def convertir_coordenadas_a_decimal(coordenada):
    # Expresin regular para extraer los valores numricos
    patron = r"([+-]?\d+(?:\.\d+)?)"
    valores = re.findall(patron, coordenada)

    # Extraer los componentes de grados, minutos y segundos
    grados = float(valores[0])

    # Verificar si la coordenada es hacia el sur o hacia el oeste
    if "S" in coordenada or "W" in coordenada:
        grados = -grados

    minutos = float(valores[1])
    segundos = float(valores[2])

    # Calcular el valor decimal
    valor_decimal = grados + minutos/60 + segundos/3600

    return valor_decimal

# Crear una funcin vectorizada
convertir_coordenadas_vectorizada = np.vectorize(convertir_coordenadas_a_decimal)

# Aplicar la funcin vectorizada a la matriz de coordenadas
coordenadas_decimales_1 = convertir_coordenadas_vectorizada(latitud_serie)

# Aplicar la funcin vectorizada a la matriz de coordenadas
coordenadas_decimales_2 = convertir_coordenadas_vectorizada(longitud_serie)
```
